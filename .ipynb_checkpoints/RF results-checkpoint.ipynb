```python
import xarray as xr
import pandas as pd
import iris
import numpy as np
import sklearn
import cartopy.crs as ccrs
from matplotlib import pyplot as plt
import copy

%matplotlib inline
```

```python
from sklearn.externals import joblib
models = {}
for i in [2,4,8]:
    models[i] = joblib.load('models/RF{:d}.pkl'.format(i))
```

```python
coord_system = iris.coord_systems.RotatedGeogCS(37.5, 177.5, 
                                                ellipsoid=iris.coord_systems.GeogCS(6371229.0)).as_cartopy_crs()

def process(tx):
    tx = tx.apply(lambda x: xr.DataArray(x.data.reshape((187, 48, 419)), dims=['time', 'y', 'x']))
    proj = coord_system.transform_points(ccrs.PlateCarree(), 
                                         tx['longitude'].data.flatten(), 
                                         tx['latitude'].data.flatten()
                                        ).reshape(187, 48, 419, 3)[:,:,:,:2]
    tx['lon'] = (('time', 'y', 'x'), proj[:,:,:,0])
    tx['lat'] = (('time', 'y', 'x'), proj[:,:,:,1])
    return tx
```

```python
test_sets = {}
for i in [2,4,8]:
    tdf = pd.read_csv('../data/id_test{:d}.csv'.format(i))
    tdf = tdf[tdf['fcst_id'] > 0]
    tx = xr.Dataset(tdf)
    tx = process(tx)
    test_sets[i] = tx
```

```python
def predict(model, test):
    skip_vars = ['Unnamed: 0', 'Unnamed: 0.1', 'target', 'proj', 'lat', 'lon', 'fcst_id']
    log_vars = ['stratiform_rainfall_amount', 'stratiform_rainfall_amount_up',
         'stratiform_rainfall_amount_down', 'stratiform_rainfall_amount_left', 'stratiform_rainfall_amount_right'] 

    Xt = test[[k for k in test.data_vars.keys() if k not in skip_vars]]
    for v in log_vars:
        Xt[v] = np.log1p(Xt[v])
    preds = model.predict(Xt.to_array().data.reshape((15, -1)).T)
    return xr.DataArray(np.expm1(preds).reshape(187, 48, 419), dims=('time', 'y', 'x'))
```

```python
preds = {i: predict(models[i], test_sets[i]) for i in [2,4,8]}
```

```python
errors = {i: preds[i] - test_sets[i]['target'] for i in [2,4,8]}
```

```python
baseline_errors = {i: test_sets[i]['stratiform_rainfall_amount'] - test_sets[i]['target'] for i in [2,4,8]}
```

```python
[np.mean(np.square(e.data)) for e in errors.values()]
```

```python
[np.mean(np.square(e.data)) for e in baseline_errors.values()]
```

```python
import gc
partial_errs = {}
for i in [2,4,8]:
    partial_errs[i] = []
    for n_trees in range(2, 20):
        errs = []
        for _ in range(6):
            m = copy.copy(models[i])
            ind = np.random.choice(len(m.estimators_), size=n_trees, replace=False)
            m.estimators_ = list(np.array(m.estimators_)[ind])
            e = np.mean(np.square(predict(m, test_sets[i]).data - test_sets[i]['target'].data))
            errs.append(e)
            gc.collect(2)
        partial_errs[i].append(np.mean(errs))
```

```python
for i, e in partial_errs.items():
    plt.plot(e, label=i)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 1))
plt.title('rainfall')
plt.plot(np.mean(test_sets[2]['target'], axis=(1,2)))
plt.gca().set_xticks(range(0,190,5))
plt.show()
for i in [2,4,8]:
    plt.figure(figsize=(15, 1))
    plt.title('mse {:d}'.format(i))
    plt.plot(np.mean(np.square(errors[i].data), axis=(1,2)), label='model')
    plt.plot(np.mean(np.square(baseline_errors[i].data), axis=(1,2)), label='baseline')
    plt.gca().set_xticks(range(0,190,5))
    plt.show()
```

```python
f, ax = plt.subplots(1, 3, figsize=(12, 4))
for j, i in enumerate([2,4,8]):
    yhat = preds[i].data
    true = (test_sets[i]['target'].data > 0).astype(int)
    fpr, tpr, _ = sklearn.metrics.roc_curve(true.flatten(), yhat.flatten())
    ax[j].plot(fpr, tpr, label='model')
    ax[j].set_title(i)
    ax[j].set_xlabel('False Positive Rate')
    ax[j].set_ylabel('True Positive Rate')
    
    yhat = test_sets[i]['stratiform_rainfall_amount'].data
    true = (test_sets[i]['target'].data > 0).astype(int)
    fpr, tpr, _ = sklearn.metrics.roc_curve(true.flatten(), yhat.flatten())
    ax[j].plot(fpr, tpr, label='baseline')
    ax[j].set_title(i)
    ax[j].set_xlabel('False Positive Rate')
    ax[j].set_ylabel('True Positive Rate')
plt.legend()
plt.tight_layout()
```
